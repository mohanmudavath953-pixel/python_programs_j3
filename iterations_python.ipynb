# iterations in python 
1.iterations means loops to work with repeatative task or repeatative performances
2. python have 2 types of iterators
3.`for`loop
4.`while`loop


```python - for loop 
for iter_variable in sequence:
    print("for is keyword to start the for loop")
    print("iter_variable is one type of variable to iterate data from sequence")
    print("in is keyword to validate sequence is present or not")
    print("sequence is collection like range,list,string,tuple,...etc")
```

```python - while loop 
while(condition):
    print("while is keyword to start the iteration based on condition")
    print("condition is true then executed inside statements")
```

````python
#write a program to print 1 to 50 number using for loop
for num in range(1,51):
    print(num)
````
