# dict in data structures 
1. dict is collection of items enclosed by {} - here item is a key value pair

```python
# create dict 
info={"name":"shiva","location":"hyderbad","salary":15000}
print(info)
print(type(info))
```

```python
#method or functions
print(dir(dict))
```

```python
#keys,values,items,get
info={"name":"shiva","location":"hyderbad","salary":15000}
print(info)
print(info.keys())
print(info.values())
print(info.items())
print(info.get("location"))
```

```python
# update()
info={"name":"shiva","location":"hyderbad","salary":15000}
print(info)
info.update({"name":"vikram"})
print(info)
info.update({"domain":"data engineer"})
print(info)
```

```python
#pop()
info={"name":"shiva","location":"hyderbad","salary":15000}
print(info)
info.pop("location")
print(info)
```

# sets in a data structure 
1.set is a collection of unorder unquie element enclosed by{}

```python
# create set
a={1,2,3,4,5,6,7,8,9,1,8,2,2,5,5,2,2,2,2,}
print(a)
print(type(a))
```

```python
# methods or function in set
print(dir(set))
```

```python
#odd()
a={10,30,50,40,}
print(a)
a.add(100)
print(a)
```

```python
#difference 
a=(1,2,3,4,5)
b={3,4,5,6,7,8}
c=b.difference(a)
print(c)
```

```python
#union
a={1,2,3,4,5}
b={3,4,5,6,7,8}
c=a.union(b)
print(c)
```
