```python
#repeat string
name="Machine Learning"
print(name*100)
```

```python
### accessing characters from string - index based
#### two days 
1. forward index- read characters from left right(0,1,2,3,...n-1)
2. backward or negative index- read characters from right to left (-1,-2,-3,......-n)
```

```python
name="machine learning"
print(name)
print(name[0])
```

```python
name="machine Learning"
print(name)
print(name[8])
```

```python
#silicing concept - accesing group of characters - string[start_index:end_index-1]
name="machine learning"
print(name)
print(name[0:7])
```
